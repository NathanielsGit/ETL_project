```python
# Import libraries and dependencies
import pandas as pd
import sqlite3
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import Table, Column, Integer, String, MetaData, ForeignKey
from sqlalchemy import inspect
```

```python
# Saving file paths as variables and loading the datasets
books = "Resources/csv/books.csv"
books_df = pd.read_csv(books, encoding="ISO-8859-1")
books_df.head(2)
```

```python
#Renaming columns for readability
books_df = books_df.rename(columns={'Book-Title': 'Title', 'Book-Author': 'Author', 'Year-Of-Publication': 'Publication'})
books_df.head(2)
```

```python
#Keeping and re-arranging necessary columns 
books_df = books_df[['ISBN', 'Title', 'Author', 'Publication','Publisher']]
books_df.head(2)
```

```python
books_df.nunique()
```

```python
#Deleting Null value rows

nan_value = float("NaN")
books_df.replace("", nan_value, inplace=True)
books_df.dropna(subset = ['ISBN', 'Title', 'Author', 'Publication','Publisher'], inplace=True)
books_df
```

```python
# Saving file path as variable and loading the 'ratings' dataset
ratings = "Resources/csv/ratings.csv"
ratings_df = pd.read_csv(ratings, encoding="ISO-8859-1")
ratings_df.head(2)
```

```python
#Renaming columns for readability
ratings_df = ratings_df.rename(columns={'User-ID': 'User_ID', 'Book-Rating': 'Rating'})
ratings_df.head()
```

```python
#Keeping and re-arranging necessary columns 
ratings_df = ratings_df[['ISBN', 'User_ID' ,'Rating']]
ratings_df.head()
```

```python
#Deleting Null value rows
nan_value = float("NaN")
ratings_df.replace("", nan_value, inplace=True)
ratings_df.dropna(subset = ['ISBN', 'User_ID' ,'Rating'], inplace=True)
ratings_df
```

```python
# Save file paths and load datasets as variables
users = "Resources/csv/users.csv"
users_df = pd.read_csv(users, encoding="ISO-8859-1")
users_df.head()
```

```python
#Rename and rearrange columns 
users_df = users_df.rename(columns={'User-ID': 'User_ID'})
users_df.head()
```

```python
#Deleting Null value rows
nan_value = float("NaN")
users_df.replace("", nan_value, inplace=True)
users_df.dropna(subset = ['User_ID', 'City', 'Country', 'Age'], inplace=True)
users_df.head()
```

```python
#merging the books and ratings data frames

books_ratings_df = pd.merge(left=books_df, right= ratings_df, left_on="ISBN", right_on="ISBN", how="inner")

books_ratings_df.head()
```

```python
#merging the books_ratings data frame with users
final_df = pd.merge(left=books_ratings_df, right=users_df, left_on="User_ID", right_on="User_ID", how="left")
final_df.head()
```

```python
#Rearranging columns for readability
final_df = final_df[['User_ID', 'City', 'Country', 'Age','ISBN', 'Title', 'Author', 'Publication','Publisher', 'Rating']]
final_df.head()
```

```python
#Deleting Null values and empty strings
nan_value = float("NaN")
final_df.replace("", nan_value, inplace=True)
final_df.dropna(subset = ['User_ID', 'City', 'Country', 'Age','ISBN', 'Title', 'Author', 'Publication','Publisher', 'Rating'], inplace=True)
final_df.head()
```

```python
#Save 'final_df' to csv
final_df.to_csv('Resources/clean/Clean_CSV')
```
